# ETS Decomposition (Error + Trend + Seasonality)

```python
# import packages 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (12,6)
sns.set()
```

```python
df = pd.read_csv("./data/airline_passengers.csv", index_col="Month")
df.head()
```

```python
sns.heatmap(df.isnull())
```

```python
df.dropna(inplace=True)
```

```python
_ = df.plot()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
airline_decomposed = seasonal_decompose(df["Thousands of Passengers"], model="multiplicative")
```

```python
airline_decomposed.seasonal.plot()
```

```python
airline_decomposed.trend.plot()
```

```python
_ = airline_decomposed.plot()
```
